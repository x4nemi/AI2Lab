# Regresión Lineal Univariable en Python

**Intrucciones**: configura los parámetros adecuados como se indica en las instrucciones del reporte para que el código funcione.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('dataset_RegresionLineal.csv')
x = np.array(data['x'])
y = np.array(data['y'])
m = np.size(x)
```

## Graficar datos

```python
plt.plot(x,y,'o',color = "blue",mec = "black")
plt.title('Gráfica de dispersión')
plt.xlabel('x: área en m²')
plt.ylabel('y: costo')
```

## Parámetros

```python
a0 = 0
a1 = 0
beta = 0.023
iterMax = 600
iter = 1
```

## Calcular hipótesis con parámetros iniciales

```python
h = a0 + a1*x
plt.figure(1)
plt.plot(x, h, 'r')
```

## Calcular función de costo $J$

```python
J = (1/(2*m))*sum(np.power((h-y),2))
convergencia=np.zeros(iterMax)
convergencia[0]=J
```

## Entrenamiento

```python
while iter < iterMax:
    a0 = a0 - beta*(1/m)*sum(h-y)
    a1 = a1 -  beta*(1/m)*sum((h-y)*x)
    h = a0 + a1*x
    #plt.plot(x,h,'y')
    J = (1/(2*m))*sum(np.power((h-y),2))
    convergencia[iter]=J
    iter += 1
```

## Dibujar el modelo final

```python
plt.figure(2)
plt.plot(convergencia)
plt.title('Grafica de convergencia')
plt.xlabel('iteraciones')
plt.ylabel('J')
```

## Imprimir resultados finales

```python
print('J =',J)
print('a0 =',a0,' a1 =',a1)
```

## Hacer una predicción usando el modelo

```python
datoPrueba = 83
h_prueba = a0 + a1*x[datoPrueba]
plt.figure(1)
plt.plot(x[datoPrueba],h_prueba,'x',color='black',mec='black')
print("Dato prueba = ",x[datoPrueba],'Salida correcta = ',y[datoPrueba],'Predicción=',h_prueba)
```

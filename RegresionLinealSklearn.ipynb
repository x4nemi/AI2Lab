# Regresi√≥n Lineal Univariable con `sklearn`

### Datos de entrada

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model

data = pd.read_csv('dataset_RegresionLineal.csv')
x = np.array(data['x'])
y = np.array(data['y'])

plt.plot(x, y, 'o', color="magenta", mec="black")
plt.xlabel('x')
plt.ylabel('y')
```

### Modelo de regresi√≥n lineal univariable por sklearn

```python
model = linear_model.LinearRegression()
model.fit(x.reshape(-1, 1), y)

h = model.predict(x.reshape(-1, 1))
plt.plot(x, y, 'o', color="magenta", mec="black")
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, h, color="blue")
print('a0:', model.intercept_, 'a1:', model.coef_[0])
```

### Evaluaci√≥n del modelo con un dato de prueba

```python
datoPrueba = 5.62
prediccion = model.predict([[datoPrueba]])
print('Prediccion para x =', datoPrueba, 'es', prediccion[0])

plt.plot(x, y, 'o', color="magenta", mec="black")
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, h, color="blue")
plt.plot(datoPrueba, prediccion, 'o', mec = "black", color="yellow", markersize=10)
```
